```python
#Построить график функции: y =|{x} − 1/2| = |2x-1| /2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
fig, ax = plt.subplots()
x = np.linspace(-2, 5, 200)
y = abs(2*x-1)/2
ax.plot(x,y)
plt.show()
```
